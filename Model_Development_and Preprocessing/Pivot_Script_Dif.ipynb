```python
import pandas as pd
from prophet import Prophet
import pickle

df_train = df_long[(df_long['Date'].dt.year >= 1968) & (df_long['Date'].dt.year <= 2020)]
df_prophet = df_train[['Date', 'Radiation']].rename(columns={'Date': 'ds', 'Radiation': 'y'})

model = Prophet(yearly_seasonality=True)
model.fit(df_prophet)

with open('Diff_radiation_model.pkl', 'wb') as f:
    pickle.dump(model, f)

print("✅ Model trained and saved as 'Diff_radiation_model.pkl'")

with open('Diff_radiation_model.pkl', 'rb') as f:
    model = pickle.load(f)

def forecast_for_year(model, year: int):
    dates = pd.date_range(f"{year}-01-01", f"{year}-12-31", freq='D')
    future_df = pd.DataFrame({'ds': dates})
    forecast = model.predict(future_df)
    forecast = forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']]
    forecast.columns = ['Date', 'Forecast_Radiation', 'Lower_Bound', 'Upper_Bound']
    return forecast

year = 2024
forecast_yr_df = forecast_for_year(model, year)
forecast_yr_df.to_csv(f"radiation_forecast_{year}.csv", index=False)

with open('Diff_radiation_model.pkl', 'rb') as f:
    model = pickle.load(f)

def forecast_for_date(model, date: str):
    future_df = pd.DataFrame({'ds': [pd.to_datetime(date)]})
    forecast = model.predict(future_df)
    row = forecast.iloc[0]
    return {
        'Date': row['ds'].strftime('%Y-%m-%d'),
        'Forecast_Radiation': row['yhat'],
        'Lower_Bound': row['yhat_lower'],
        'Upper_Bound': row['yhat_upper']
    }

date = "2024-06-06"
forecast_date_result = forecast_for_date(model, date)
print("🔎 Radiation forecast on", date, "→", forecast_date_result)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from prophet import Prophet
from prophet.diagnostics import cross_validation, performance_metrics
from prophet.plot import plot_cross_validation_metric

# 1. Prepare the dataset
df_prophet = (
    df_long[["Date", "Radiation"]]
    .rename(columns={"Date": "ds", "Radiation": "y"})
    .sort_values("ds")
    .reset_index(drop=True)
)

# 2. Fit the Prophet model
model = Prophet(
    daily_seasonality=False,
    weekly_seasonality=False,
    yearly_seasonality=False,
    changepoint_prior_scale=0.05,
    seasonality_prior_scale=10.0,
    interval_width=0.95
)

# Add custom seasonalities
model.add_seasonality(name="weekly", period=7, fourier_order=10)
model.add_seasonality(name="yearly", period=365.25, fourier_order=20)

model.fit(df_prophet)

# 3. Forecast next 365 days
future = model.make_future_dataframe(periods=365, freq="D")
forecast = model.predict(future)

# 4. Plot the forecast
fig1 = model.plot(forecast)
plt.title("Radiation Forecast with Prophet")
plt.xlabel("Date")
plt.ylabel("Radiation")
plt.tight_layout()
plt.show()

# 5. Cross-validation to get RMSE
df_cv = cross_validation(
    model,
    initial="1095 days",  # 3 years
    period="180 days",    # every 180 days
    horizon="180 days",   # predict next 180 days
    parallel="processes"
)

# 6. Performance metrics
df_p = performance_metrics(df_cv)
print(df_p[["horizon", "rmse"]].head())

# 7. RMSE Plot
fig2 = plot_cross_validation_metric(df_cv, metric="rmse")
plt.title("Prophet CV RMSE over Forecast Horizon")
plt.xlabel("Horizon")
plt.ylabel("RMSE")
plt.tight_layout()
plt.show()
```
