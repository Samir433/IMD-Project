```python
df_corrected = daily_good_df.copy()
prev_day = df_corrected.shift(1)
next_day = df_corrected.shift(-1)

mask_missing = df_corrected.isna() | (df_corrected == -1)
mask_valid_neighbors = (prev_day.notna()) & (prev_day >= 0) & (next_day.notna()) & (next_day >= 0)

mask_to_replace = mask_missing & mask_valid_neighbors
df_corrected[mask_to_replace] = (prev_day[mask_to_replace] + next_day[mask_to_replace]) / 2
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from prophet import Prophet
from prophet.diagnostics import cross_validation, performance_metrics

df_prophet = (
    df_long[["Date", "Radiation"]]
    .rename(columns={"Date": "ds", "Radiation": "y"})
    .sort_values("ds")
    .reset_index(drop=True)
)

model = Prophet(
    daily_seasonality=False,     # Turn off automatic (we’ll add manually if needed)
    weekly_seasonality=False,
    yearly_seasonality=False,    # Turn off automatic; we’ll add a custom one
    changepoint_prior_scale=0.05, # Default is 0.05; increase to allow more trend flexibility (try 0.1 or 0.2 if needed)
    seasonality_prior_scale=10.0, # Increase seasonal flexibility (default is 10). Tweak to reduce overfitting.
    interval_width=0.95          # 95% prediction intervals
)

model.add_seasonality(name="weekly", period=7, fourier_order=10)
model.add_seasonality(name="yearly", period=365.25, fourier_order=20)

model.fit(df_prophet)

future = model.make_future_dataframe(periods=365, freq="D")

# If you included regressors, you must add them into 'future' as well.
# For example, if you have future temperature projections or wish to fill with NaN:
# future = future.merge(df_temp_projection, on="ds", how="left")

forecast = model.predict(future)

# -----------------------------------------------------------
# 6. PLOT RESULTS
# -----------------------------------------------------------
fig1 = model.plot(forecast)
plt.title("Radiation Forecast with Prophet")
plt.xlabel("Date")
plt.ylabel("Radiation")
plt.tight_layout()

fig2 = model.plot_components(forecast)
plt.tight_layout()

plt.show()

# -----------------------------------------------------------
# 7. CROSS-VALIDATION & PERFORMANCE METRICS (OPTIONAL, BUT RECOMMENDED)
# -----------------------------------------------------------
# This lets you see how well Prophet is doing and compare parameter choices.

# 7.a. Perform cross-validation with an initial 3-year training, a 180-day horizon, and 90-day period
# (Adjust initial/horizon/period depending on your data history length)
df_cv = cross_validation(
    model,
    initial="1095 days",  # first 3 years used to train
    period="180 days",    # make predictions every 180 days
    horizon="180 days",   # for each cutoff, evaluate 180 days ahead
    parallel="processes", # use all CPU cores to speed up
)

# 7.b. Compute MAPE, RMSE, MAE, etc.
df_p = performance_metrics(df_cv)
print(df_p.head())


from prophet.plot import plot_cross_validation_metric

fig3 = plot_cross_validation_metric(df_cv, metric="rmse")
plt.title("Prophet CV RMSE over Forecast Horizon")
plt.xlabel("Horizon")
plt.ylabel("RMSE")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from prophet import Prophet
import pickle

# ── (A) LOAD THE PICKLED MODEL ──
with open('radiation_model.pkl', 'rb') as f:
    model = pickle.load(f)

# ── (B) FUNCTION TO FORECAST A FULL YEAR ──
def forecast_full_year(model, input_year: int):
    # Build one row per day in that year
    all_dates = pd.date_range(
        start=pd.Timestamp(f"{input_year}-01-01"),
        end=pd.Timestamp(f"{input_year}-12-31"),
        freq='D'
    )
    df_future = pd.DataFrame({'ds': all_dates})
    forecast_df = model.predict(df_future)
    out = forecast_df[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].copy()
    out.columns = ['Date', 'Forecast_Radiation', 'Lower_Bound', 'Upper_Bound']
    return out

# ── (C) EXAMPLE USAGE ──
year_requested = 2024
df_result = forecast_full_year(model, year_requested)

# Save to CSV (optional)
df_result.to_csv(f"radiation_forecast_{year_requested}.csv", index=False)
print(f"✅ Forecast for {year_requested} written to 'radiation_forecast_{year_requested}.csv'")
```

```python
import pandas as pd
from prophet import Prophet
import pickle
import sys

df_train = df_long[
    (df_long['Date'].dt.year >= 1968) &
    (df_long['Date'].dt.year <= 2020)
].copy()

df_prophet = df_train[['Date', 'Radiation']].rename(columns={'Date': 'ds', 'Radiation': 'y'})

model = Prophet(yearly_seasonality=True)
model.fit(df_prophet)

pickle_filename = "global_radiation_model_single_date.pkl"
with open(pickle_filename, 'wb') as f:
    pickle.dump(model, f)
print(f"✅ Model trained and saved to '{pickle_filename}'.")

with open(pickle_filename, 'rb') as f:
    loaded_model = pickle.load(f)
```
